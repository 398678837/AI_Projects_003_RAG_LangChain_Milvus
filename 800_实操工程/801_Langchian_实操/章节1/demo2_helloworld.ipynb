```python
# 导入 dotenv 库，用于加载环境变量文件(.env)中的配置
import dotenv
from langchain_openai import ChatOpenAI
import os

# 加载当前目录下的 .env 文件
dotenv.load_dotenv()

# 从环境变量读取配置
os.environ['OPENAI_API_KEY'] = os.getenv("minimax_api_key")

# 创建大模型实例，使用 openai_api_base 参数指定自定义 API 地址
llm = ChatOpenAI(
    model="MiniMax-M2",
    openai_api_base="https://api.minimaxi.com/v1",
    openai_api_key=os.getenv("minimax_api_key")
)

# 直接提问并调用模型
response = llm.invoke("什么是大模型?")
print(response.content.encode('utf-8').decode('utf-8'))
```

```python
# 2. 使用提示词模板

from langchain_core.prompts import ChatPromptTemplate

# 定义模板：system 角色 + user 变量输入
prompt = ChatPromptTemplate.from_messages([
    ("system", "你是世界级的技术文档编写者"),
    ("user", "{input}")  # {input} 是变量
])

# 3. 把模板和LLM拼在一起（LangChain 最核心写法：| 管道）
chain = prompt | llm

# 4. 调用
message = chain.invoke({"input":"大模型中的LangChain是什么?"})

# 5. 输出干净内容（清洗！只打印文字，不打印对象）
print(message)
```

```python
# 3. 使用输出解析器

# 导入包
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser, JsonOutputParser

# 创建提示模板（顺序必须是 system 在前！）
prompt = ChatPromptTemplate.from_messages([
    ("system", "你是世界级的技术文档编写者。"),
    ("user", "{input}")
])

# 输出解析器：JSON格式
output_parser = JsonOutputParser()

# 构建链：提示词 → LLM → 解析器
chain = prompt | llm | output_parser

# 调用（要求返回JSON）
result = chain.invoke({
    "input": "LangChain是什么? 用JSON格式回复，字段是 question 和 answer"
})

# 输出结果
print(result)
```

```python
#  4. 使用向量存储

# =========================
# 1. 安装依赖（运行一次即可）
# =========================
# pip install faiss-cpu
# pip install langchain_community bs4 langchain_openai

# =========================
# 2. 导入所有需要的包
# =========================
import bs4
import os
import dotenv
from langchain_community.document_loaders import WebBaseLoader
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter

# 加载环境变量
dotenv.load_dotenv()
os.environ['OPENAI_API_KEY'] = os.getenv("minimax_api_key")

# =========================
# 3. 加载网页文档
# =========================
loader = WebBaseLoader(
    web_path="https://www.gov.cn/lianbo/202604/content_7064530.htm",
    bs_kwargs=dict(parse_only=bs4.SoupStrainer(id="UCAP-CONTENT"))
)

docs = loader.load()

# =========================
# 4. 文档切分（分块）
# =========================
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)
documents = text_splitter.split_documents(docs)
print("切分后的文档数量：", len(documents))

# =========================
# 5. 初始化向量模型（嵌入模型）
# =========================
embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)

# =========================
# 6. 创建 FAISS 向量库 ✅
# =========================
vector = FAISS.from_documents(documents, embeddings)

print("✅ FAISS 向量库创建成功！")
```

```python
from langchain_core.prompts import PromptTemplate

retriever = vector.as_retriever()
retriever.search_kwargs = {"k": 3}
docs = retriever.invoke("建设用地使用权是什么？")

# for i,doc in enumerate(docs):
#     print(f"⭐第{i+1}条规定：")
#     print(doc)

# 6.定义提示词模版
prompt_template = """
你是一个问答机器人。
你的任务是根据下述给定的已知信息回答用户问题。
确保你的回复完全依据下述已知信息。不要编造答案。
如果下述已知信息不足以回答用户的问题，请直接回复"我无法回答您的问题"。

已知信息:
{info}

用户问：
{question}

请用中文回答用户问题。
"""
# 7.得到提示词模版对象
template = PromptTemplate.from_template(prompt_template)

# 8.得到提示词对象
prompt = template.format(info=docs, question='一起益企是什么？')

## 9. 调用LLM
response = llm.invoke(prompt)
print(response.content)
```

```python
# 6.使用agent
from langchain_community.tools import Tool

# 手动创建检索器工具
def retriever_tool_func(query):
    docs = retriever.get_relevant_documents(query)
    return "

".join([doc.page_content for doc in docs])

retriever_tool = Tool(
    name="CivilCodeRetriever",
    func=retriever_tool_func,
    description="搜索有关中华人民共和国民法典的信息。关于中华人民共和国民法典的任何问题，您必须使用此工具!"
)

tools = [retriever_tool]

from langchain import hub
from langchain.agents import create_openai_functions_agent
from langchain.agents import AgentExecutor

# https://smith.langchain.com/hub
prompt = hub.pull("hwchase17/openai-functions-agent")

agent = create_openai_functions_agent(llm, tools, prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# 运行代理
agent_executor.invoke({"input": "一起益企是什么？"})
```
